```python
#QUESTION 3:
#plotting various histograms
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
housing = pd.read_csv("housing.csv")
housing.head()
#filtered data and saved into new data frame on condition median house value > 50000 and  housing median age < 50.
df_housing = housing[  (housing['median_house_value'] > 50000 )& (housing['housing_median_age'] < 50) ]
df_housing
#plotted histogram for the housing data
df_housing.hist(bins=50,figsize=(20,15))
```
